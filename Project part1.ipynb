```python
#importing the relevant libraries for data visualization
import numpy as np                #importing numpy(numerical processing library)
import pandas as pd               #importing pandas
import matplotlib.pyplot as plt   #importing pyplot module from matplotlib library
import seaborn as sns             #importing seaborn
```

__1. Reading the data from the CSV file into a pandas DataFrame__

```python
df=pd.read_csv("C:\\Users\\dhruv\\Downloads\\NDAP_REPORT_7004.csv")    #Reading the data from the given csv file into a pandas dataframe
display(df)                                                            #Displaying the data
```

__2. Print the number of columns and rows__

```python
shape=df.shape                                                            #this method return a tuple in the form (rows,columns)
print("Number of columns in the DataFrame is {}".format(shape[1]))        #obtaining the number of columns using indexing
print("Number of rows int the DataFrame is {}".format(shape[0]))          #obtaining the number of rows using indexing
```

__3. Print Number of unique values in each column__

```python
for columns in df.columns.values:                                         #running a for loop to iterate through the columns of the dataframe
    unique_values=pd.unique(df[columns])                                  #creating a python list which contains all the unique elements in a columns
    print("{} has {} unique values".format(columns,len(unique_values)))   #length of the unique_vals list = number of unique values in a column
```

__4. Print DataType of each column__

```python
for columns in df.columns.values:                                         #running a for loop to iterate through the columns if the dataframe
    print("{} has Data Type --> {}".format(columns,df[columns].dtype))    #.dtype method return the datatype of the column specified int the []
```

__5. Plot a Histogram of the number of null values in each column in ascending order__

```python
null_list=[]
for columns in df.columns.values:
    null_list.append(np.sum(pd.isnull(df[columns])))                       #creating a list which contains the number of null values in every columns
null_frame=pd.DataFrame({"Columns":df.columns,"Null Values":null_list})    #creating a dataframe from the null list
null_frame=null_frame.sort_values("Null Values",ascending=True)            #sorting the values in ascending order
plt.figure(figsize=(10,10))                                                #used to change the size of the plot
plt.bar(null_frame["Columns"],null_frame["Null Values"],color="maroon",width=0.6)
plt.yticks(np.arange(min(null_list),7,1))                                  #specifying the yticks
plt.xticks(rotation=90)                                                    #rotating the xticks by 90 degrees so that they do not overlap
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.title("Number of Null Values in each column")
plt.show()
```

__6. Plot a Histogram of the number of null values in each row in ascending order__

```python
null_list=[]                                                              #creating a list to store the null of null values of every row in the dataframe
for index in df.index:
    null_list.append(np.sum(pd.isnull(df.iloc[index])))
null_frame=pd.DataFrame({"Rows":df.index,"Null Values":null_list})        #creating a dataframe in order to plot the barplot
null_frame=null_frame.sort_values("Null Values")
display(null_frame)
plt.figure(figsize=(40,10))
plt.bar(null_frame["Rows"],null_frame["Null Values"],color="maroon")
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```

__7. Data Cleaning__

```python
#for data cleaning we need to replace the "zero" by 0(integer) wherever it occurs and drop the null/nan values
#only the columns listed below have the value '0' as "zero"
df["Number of census houses"]=pd.to_numeric(df.replace({"Number of census houses":{"zero":0}})["Number of census houses"])
df["Houses having grass or thatch or bamboo etc. as material of wall"]=pd.to_numeric(df.replace({"Houses having grass or thatch or bamboo etc. as material of wall":{"zero":0}})["Houses having grass or thatch or bamboo etc. as material of wall"])
df["Houses having plastic or polythene as material of wall"]=pd.to_numeric(df.replace({"Houses having plastic or polythene as material of wall":{"zero":0}})["Houses having plastic or polythene as material of wall"])
for column in df.columns:
  df[column] = df[column].dropna()
```

```python
#printing the data types of each column to check for any mistakes
print(df.dtypes)
#everyting seems fine.
```

__8. For all Discrete Variables, plot Histograms__

```python
#creating a copy of the original dataframe in order to preserve the data in the original dataframe
temp_df=df
#dropping the columns for which, plotting a graph does not yield any useful information
#basically, it is useless to plot for these columns
temp_df=temp_df.drop("Rowid",axis=1)
temp_df=temp_df.drop("Year",axis=1)
temp_df=temp_df.drop("District",axis=1)
temp_df=temp_df.drop("State lgd code",axis=1)
temp_df=temp_df.drop("District lgd code",axis=1)
temp_df=temp_df.drop("Country",axis=1)
for col in temp_df.columns.values:
  if (temp_df[col].dtypes == 'int64') | (temp_df[col].dtypes == 'object'):
    if(col=="State"):                                   #I have specified this case discretely because in this case, the xticks were overlapping
        plt.figure(figsize=(10,10))
        plt.hist(temp_df[col],color="maroon")
        plt.xticks(rotation=90,fontsize=12)
        plt.yticks(fontsize=12)
        plt.ylabel("Number of districts in each state",fontsize=15)
        plt.xlabel(col,fontsize=15)
        plt.show()
    else:   
        plt.figure(figsize=(10,10))
        plt.hist(temp_df[col],color="maroon")
        plt.xticks(fontsize=12)
        plt.yticks(fontsize=12)
        plt.xlabel(col,fontsize=15)
        plt.show()
```

__9. Plotting histogram for the continuous variables using appropriate number of bins__

```python
for column in df.columns.values:                               #using a for loop to iterate through the columns
    if(df[column].dtype=='float64'):
        plt.figure(figsize=(10,10))
        plt.hist(df[column],15 if len(pd.unique(df[column]))>20 else len(pd.unique(df[column])),color="maroon")    #using the appropriate number if bins
        plt.xlabel(column,fontsize=15)                         #specifying the xlabel
        plt.xticks(fontsize=12)
        plt.yticks(fontsize=12)
        plt.show()
```

__10. For all pairs of continuous variables, plot the scatter plot and show the color-coded correlation matrix__

```python
from pandas.plotting import scatter_matrix
continuous_df=df
#creating a copy of the original dataframe so that no changes are made to the original dataframe
#dropping all the discrete variables because we have to plot for all the continuous variables
continuous_df=continuous_df.drop("Rowid",axis=1)
continuous_df=continuous_df.drop("Country",axis=1)
continuous_df=continuous_df.drop("State lgd code",axis=1)
continuous_df=continuous_df.drop("State",axis=1)
continuous_df=continuous_df.drop("District lgd code",axis=1)
continuous_df=continuous_df.drop("District",axis=1)
continuous_df=continuous_df.drop("Year",axis=1)
continuous_df=continuous_df.drop("Type of residence namely rural or urban",axis=1)
scatter_matrix(continuous_df, alpha=0.2, figsize=(26, 26), diagonal='kde')          #plotting the scatter matrix
display(continuous_df)                                                              #displaying the copy dataframe to check whether the unnecessary columns have been dropped or not
```

```python
#Plotting the color-coded correlation matrix
corrMatrix = continuous_df.corr(method='pearson') # Use spearman for non-linear
fig, ax = plt.subplots(figsize=(20,20))
sns.heatmap(abs(corrMatrix), annot=False) # Show absolute value
plt.xticks(fontsize=15)                   #setting the fontsize of the xticks
plt.yticks(fontsize=15)                   #setting the fontsize of the yticks
plt.show()
```

__11. Check if the sum of columns J through S matches with column I for all rows__

```python
new_df = df.iloc[:, 9:19]                 #creating a new dataframe which only contains the columns J through S
# new_df=new_df.replace([np.inf,-np.inf],np.nan).dropna()   
Sum_series = new_df.sum(axis = 1)         #creating a pandas series which contains the sum of all columns J through S
#Although, theoretically the sum of the solumns should be same as the column I for all rows but we see that it is not true for all the rows
#A possible explanation for these different values might be floating point accuracy issue due to which the sum value varies very slightly from the column I
for index in range(0,1473):
    if(Sum_series[index]==df["Number of census houses"][index]):
        print("Same value in the row {}".format(index+1))
    else:
        print("Distinct values in the row {}".format(index+1))
```

__12. Divide columns J through S by column I and store them in new columns as percent (multiply by 100)__

```python
new_df=df.iloc[:,0:9]         #defining a new dataframe containing the columns A through I.
for index in range(9,19):
    series=df.iloc[:,index].div(df["Number of census houses"])  #using .div(args) function to divide the 2 columns
    new_df[df.columns[index]]=series.multiply(other=100)        #Adding the new columns to the dataframe
    #using .multiply(args) function to multiply the newly obtained columns by 100 (for percentage)
display(new_df)
```

__13. For these new columns, show box and whiskers plot__

```python
new_df=new_df.replace([np.inf,-np.inf],np.nan).dropna()      #replacing the infinite values by null values and then dropping the null values
#Showing the box and whisker plot
for index in range(9,19):
    plt.figure(figsize=(10,10))
    plt.boxplot(new_df.iloc[:,index])                        #box and whisker plot
    plt.xlabel(new_df.columns[index],fontsize=15)
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    plt.show()
```
